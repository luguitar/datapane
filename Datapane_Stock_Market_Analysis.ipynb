# Plotagem

```python
trace0 = go.Scatter(x=petro.index, y=petro.Close, name='nflx')
fig0 = go.Figure([trace0])
fig0.update_layout(
    title={
        'text': "Ação da Petrobrás",
        'x':0.5,
        'xanchor': 'center'})
fig0.show()
```

```python
petro['10-day MA'] = petro['Close'].rolling(window=10).mean()
petro['20-day MA'] = petro['Close'].rolling(window=20).mean()
petro['50-day MA'] = petro['Close'].rolling(window=50).mean()

trace0 = go.Scatter(x=petro.index, y=petro.Close, name='Petrobrás')
trace1 = go.Scatter(x=petro.index, y=petro['10-day MA'], name='10-day MM')
trace2 = go.Scatter(x=petro.index, y=petro['20-day MA'], name='20-day MM')
fig1 = go.Figure([trace0, trace1, trace2])
fig1.update_layout(
    title={
        'text': "Ação da Petrobrás",
        'x':0.5,
        'xanchor': 'center'})
fig1.show()
```

```python
fig2 = go.Figure(go.Candlestick(x=petro.index, open=petro.Open, high=petro.High, low=petro.Low, close=petro.Close))
fig2.update_layout(
    title={
        'text': "Ação da Petrobrás (Candle Stick)",
        'x':0.5,
        'xanchor': 'center'})
fig2.show()
```

```python
petro['Retorno Diário (%)'] = round(petro['Close'].pct_change()*100, 2)

fig3 = px.bar(petro, x=petro.index, y="Retorno Diário (%)")
fig3.update_traces(marker_color='rgb(158,202,225)', marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)
fig3.update_layout(
    title={
        'text': "Retorno Diário - PETR4",
        'x':0.5,
        'xanchor': 'center'})
fig3.show()
```

```python
trace0 = go.Scatter(x=petro.index, y=petro.Close, name='Petrobrás', line=dict(color='lime'))
trace1 = go.Scatter(x=tesla.index, y=tesla.Close, name='Tesla', line=dict(color='red'))

fig4 = go.Figure([trace0, trace1])
fig4.update_layout(dict(
    title=dict({
        'text': "Petrobrás vs Tesla",
        'x':0.5,
        'xanchor': 'center'}),
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=3,
                     label='3m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(count=12,
                     label='12m',
                     step='month',
                     stepmode='backward')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date')))
fig4.show()
```

# Building a Datapane report

Now that we have a series of plots created using Plotly, we can create a report using Datapane. 

In addition to the visualizations, this report includes Datapane's `BigNumber` component to display today's stock prices, and our `DataTable` component to allow our viewers to filter, explore, and download the data themselves.

Although we could create a linear list of all blocks (similar to how they are displayed in this notebook), we can build a more powerful and accessible report by using Datapane's [layout components](https://docs.datapane.com/reports/blocks/layout-pages-and-selects). 

From these components, we are using `Group` to place the `BigNumber` blocks in two columns, have used the `Page` block to add multiple pages to our report, and are using the `Select` block to create tabs where users can choose their dataset. This results in a beautiful, interactive document which we can publish and share.


```python
import datapane as dp

petro_today = petro.iloc[-1] 
tesla_today = tesla.iloc[-1]

dp.Report(
    dp.Page(
        title='Dashboard',
        blocks=[
            dp.Group(
                dp.BigNumber(
                    heading="Petr4 Performance Diária",
                    value="${:,.2f}".format(petro_today.Close),
                    prev_value="${:,.2f}".format(petro_today.Open),
                ), 
                dp.BigNumber(
                    heading="TSLA Performance Diária",
                    value="${:,.2f}".format(tesla_today.Close),
                    prev_value="${:,.2f}".format(tesla_today.Open),
                ),        
                columns=2 
            ),
            dp.Group(
                fig0, fig1, fig2, fig3, fig4, fig5,
                columns=2, 
            ),
            fig7
      ]),
      dp.Page(title='Datasets', blocks=[
          "_These datasets are pulled live when the report is created_.",                  
          dp.Select(
              dp.DataTable(petro, label='PETR4'), 
              dp.DataTable(tesla, label='Tesla')
          )
      ])
).upload(name='Stock Market Data', open=True)
```
